# Program "ApaBole"   
  Print angka dari 1 sampai 100. Tetapi,:
1.   Untuk setiap angka yang bisa dibagi dengan 3, print "Apa". 
2.   Jika dapat dibagi dengan 5, print "Bole",
3. Jika dibagi oleh 3 dan 5, print "ApaBole".

```python
def fizbuzz(n):
    for i in range(1, n+1):
        if i % 3 == 0 and i % 5 == 0:
            print("ApaBole", end=", ")
        elif i % 3 == 0:
            print("Apa", end=", ")
        elif i % 5 == 0:
            print("Bole", end=", ")
        else:
            print(i, end=", ")

fizbuzz(100)
```

# Ramalan cuaca kota Jakarta untuk 5 hari kedepan 
Note : menggunakan API http://openweathermap.org

```python
import requests

city = "Jakarta"
api_key = "8775bf099e83fd6c1987f58bc9b257a8"

url = 'http://api.openweathermap.org/data/2.5/weather?appid='+api_key+'&q='+city

req = requests.get(url)
data = req.json()

name = data['name']
lon = data['coord']['lon']
lat = data['coord']['lat']

#print(name, "is city at atitude : ", lat, "South and longitude : ", lon, "East")
```

```python
import pandas as pd
url2 = 'http://api.openweathermap.org/data/2.5/forecast/?lat='+str(lat)+'&lon='+str(lon)+'&appid='+api_key

req2 = requests.get(url2)
data2 = req2.json()
temperature = []
date = []

for i in data2['list']:
  temperature.append(round(i['main']['temp'] - 273.15, 1))
  date.append(i['dt_txt'])

df = pd.DataFrame({'Datetime': date, 'Temperature': temperature}, columns=['Datetime', 'Temperature'])
df.head(1)
```

```python
df['Date'] = pd.to_datetime(df['Datetime']).dt.strftime('%d %B %Y')
df['Time'] = pd.to_datetime(df['Datetime']).dt.time
df['Day'] = pd.to_datetime(df['Datetime']).dt.day_name()
df['Hour'] = pd.to_datetime(df['Datetime']).dt.hour
df.head(1)
```

```python
new_cols = ["Datetime","Day","Date","Time","Hour","Temperature"]
df=df[new_cols]
df.head(1)
```

```python
Hour_0 = df.loc[df['Hour'] == 0].reset_index(drop=True)
Hour_3 = df.loc[df['Hour'] == 3].reset_index(drop=True)
Hour_6 = df.loc[df['Hour'] == 6].reset_index(drop=True)
Hour_9 = df.loc[df['Hour'] == 9].reset_index(drop=True)
Hour_12 = df.loc[df['Hour'] == 12].reset_index(drop=True)
Hour_15 = df.loc[df['Hour'] == 15].reset_index(drop=True)
Hour_18 = df.loc[df['Hour'] == 18].reset_index(drop=True)
Hour_21 = df.loc[df['Hour'] == 21].reset_index(drop=True)

Hour_0.head(1)
```

```python
def forecast(data):
  print("Weather Forecast at "+str(data['Time'][0])+" :")
  for i in data.index:
      print(data['Day'][i]+", "+str(data['Date'][i])+": "+str(data['Temperature'][i]))
  print()
```

```python
forecast(Hour_0)
forecast(Hour_3)
forecast(Hour_6)
forecast(Hour_9)
forecast(Hour_12)
forecast(Hour_15)
forecast(Hour_18)
forecast(Hour_21)
```
